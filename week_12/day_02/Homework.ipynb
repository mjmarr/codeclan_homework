## Q5

```python
drinks_grouped = drinks.groupby('beverage_category')['calories'].mean().reset_index(name = "mean_calories").copy()
drinks_grouped
```

## Q6

```python
# get average calories of all drinks
avg_all_drinks = round(drinks.calories.mean(), 2)
avg_all_drinks
```

```python
# calculate beverage_prep > average calories

drinks_prep = drinks.groupby('beverage_prep')['calories'].mean().reset_index(name = "mean_calories").copy()
drinks_prep[drinks_prep.mean_calories > avg_all_drinks]
```

## Q7

```python
# get unique beverage list
coffee_list = drinks.beverage_category.unique().tolist()
coffee_list.remove('Tazo® Tea Drinks')
coffee_list
```

```python
# filter by the list by category -> group by beverage -> get caloric values
filtered_by_cal = drinks[drinks["beverage_category"].isin(coffee_list)].groupby('beverage')['calories'].min().sort_values().reset_index().copy()

filtered_by_cal.head(10) #top 10 values
```
